# DataSet Terms Explanation

**1. Age:**  
As people get older, their risk of heart disease increases. Aging can cause changes in the heart and blood vessels, such as plaque buildup in the arteries (atherosclerosis), raising the likelihood of heart attacks and other heart problems.

**2. Gender:**  
Men generally have a higher risk of heart disease at younger ages. After menopause, women's risk increases and eventually matches that of men. Estrogen, present in women before menopause, may help protect against heart disease. Women are also more likely to have unusual symptoms of heart disease, which can sometimes delay diagnosis and treatment.

**3. Chest Pain Type:**  
Different types of chest pain can indicate different causes. Stable angina occurs during physical activity or stress and is relieved by rest. Unstable angina can happen at rest and may signal a more serious condition like a heart attack.

**4. Resting Blood Pressure:**  
High blood pressure is a major risk factor for heart disease. It can damage arteries over time, increasing the chance of plaque buildup and causing the heart to work harder, which could lead to heart failure, heart attacks, or strokes.

**5. Serum Cholesterol:**  
High levels of LDL ("bad") cholesterol and low levels of HDL ("good") cholesterol increase the risk of heart disease. LDL can form plaque in the arteries, narrowing them and restricting blood flow to the heart.

**6. Fasting Blood Sugar:**  
High fasting blood sugar can point to diabetes or prediabetes, both of which are significant risk factors for heart disease. Diabetes speeds up plaque buildup in the arteries, raising the risk of heart attacks and strokes.

**7. Resting Electrocardiographic Results (Resting ECG):**  
An abnormal ECG can indicate heart problems like coronary artery disease, arrhythmias, or heart muscle issues. Specific ECG patterns, such as ST-segment changes, can show signs of a heart attack.

**8. Maximum Heart Rate Achieved:**  
The highest heart rate reached during exercise tests can indicate cardiovascular fitness. A lower maximum heart rate could suggest reduced heart function and a higher risk of heart disease.

**9. Exercise Induced Angina:**  
Angina caused by exercise suggests coronary artery disease. It occurs when the heart doesnâ€™t get enough oxygen and blood during physical activity, signaling a potential heart attack risk.

**10. ST Depression Induced by Exercise:**  
ST-segment depression during exercise suggests reduced blood flow to the heart (myocardial ischemia). This is a strong indicator of coronary artery disease and an increased risk of heart attack.

**11. Slope of Peak Exercise ST Segment:**  
The shape of the ST segment during exercise tests gives more information about the severity of reduced blood flow to the heart. A downsloping ST segment often signals significant coronary artery disease and a higher risk of heart issues.

**12. Number of Major Vessels Colored by Fluoroscopy:**  
This measures how many major coronary arteries have significant narrowing, detected by fluoroscopy. More narrowed vessels indicate more severe coronary artery disease and a higher risk of heart attack.

**13. Thalassemia Type:**  
Thalassemia is a genetic disorder affecting hemoglobin production. Some types, particularly thalassemia major, can cause heart problems like cardiomyopathy (a disease of the heart muscle) and heart failure.

**14. Heart Disease (Target):**  
This refers to the presence or absence of heart disease, including conditions like coronary artery disease or heart attacks. The other factors listed help determine the risk of heart disease and guide diagnosis and treatment for patients.


## Causes of Heart Attack

1. **Aging**: As individuals age, the risk of developing heart-related issues increases due to changes in the heart and blood vessels.

2. **Hypertension (High Blood Pressure)**: High blood pressure damages the arteries over time, increasing the risk of atherosclerosis and heart attack.

3. **High Cholesterol Levels**: Elevated levels of LDL cholesterol ("bad" cholesterol) contribute to the formation of arterial plaques, narrowing the arteries and increasing the risk of heart attack.

4. **Diabetes**: Diabetes accelerates the development of atherosclerosis and increases the risk of heart attack and other cardiovascular complications.

5. **Obesity**: Excess body weight, particularly abdominal obesity, is associated with hypertension, diabetes, and dyslipidemia, all of which are major risk factors for heart attack.

6. **Sedentary Lifestyle**: Lack of physical activity contributes to obesity, hypertension, and dyslipidemia, increasing the risk of heart attack.

7. **Smoking**: Tobacco smoke contains chemicals that damage blood vessels and promote atherosclerosis, significantly increasing the risk of heart attack.

8. **Genetic Factors**: Family history of heart disease and genetic predispositions can increase the likelihood of developing heart attack risk factors such as hypertension, high cholesterol, and diabetes.

9. **Unhealthy Diet**: Poor dietary habits high in saturated fats, trans fats, cholesterol, and sodium contribute to hypertension, dyslipidemia, obesity, and diabetes, all of which increase heart attack risk.

10. **Stress**: Chronic stress can lead to unhealthy coping mechanisms such as overeating, smoking, and physical inactivity, exacerbating heart attack risk factors.

11. **Alcohol Consumption**: Excessive alcohol consumption can raise blood pressure, contribute to obesity, and increase triglyceride levels, all of which are associated with higher heart attack risk.

12. **Sleep Apnea**: Sleep apnea is associated with hypertension, obesity, and insulin resistance, all of which increase the risk of heart attack.

13. **Poor Dental Health**: Gum disease (periodontitis) has been linked to an increased risk of heart disease and heart attack due to the inflammation and bacterial infections it causes.

14. **Drug Abuse**: Illicit drug use, particularly stimulants like cocaine and methamphetamine, can cause arterial spasms, increased heart rate, and elevated blood pressure, raising the risk of heart attack.

15. **Undiagnosed or Untreated Heart Conditions**: Conditions such as coronary artery disease, arrhythmias, heart valve disorders, and congenital heart defects can significantly increase the risk of heart attack if left untreated or undiagnosed.

```python
def has_categorical_column(df):
    """
    Check if the DataFrame has at least one categorical column.

    Parameters:
        df (pd.DataFrame): The DataFrame to check.

    Returns:
        bool: True if there is at least one categorical column, False otherwise.
    """
    # Iterate through all columns to check for 'object' dtype
    for col in df.columns:
        if df[col].dtype == 'object':
            return True
    return False

# Example usage
if has_categorical_column(df):
    print("The DataFrame contains at least one categorical column.")
else:
    print("The DataFrame does not contain any categorical column.")
```

```python
def plot_df(df):
    """
    Generate distribution plots for numerical columns and count plots for categorical columns in a DataFrame.

    Parameters:
        df (pd.DataFrame): The DataFrame to visualize.
    """
    # Select numerical and categorical columns
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    categorical_cols = df.select_dtypes(include=['object']).columns

    # Plot histograms for numerical columns
    for col in numerical_cols:
        plt.figure(figsize=(10, 6))
        sns.histplot(df[col], kde=True, color="skyblue", bins=30)
        plt.title(f"Distribution Plot of '{col}'", fontsize=14)
        plt.xlabel(col, fontsize=12)
        plt.ylabel("Frequency", fontsize=12)
        plt.xticks(rotation=45)
        plt.grid(visible=True, linestyle='--', alpha=0.6)
        plt.show()

    # Plot count plots for categorical columns
    for col in categorical_cols:
        plt.figure(figsize=(10, 6))
        sns.countplot(x=col, data=df, palette="pastel")
        plt.title(f"Count Plot of '{col}'", fontsize=14)
        plt.xlabel(col, fontsize=12)
        plt.ylabel("Count", fontsize=12)
        plt.xticks(rotation=45)
        plt.grid(visible=True, linestyle='--', alpha=0.6)
        plt.show()

# Call the function with the DataFrame
plot_df(df)
```

```python
# Importing RobustScaler to scale features while handling outliers using median and IQR.
from sklearn.preprocessing import RobustScaler

# Splitting the dataset into features (X) and target (y).
X = df.drop('target', axis=1)  # Features
y = df['target']  # Target

# Initializing and applying RobustScaler to scale the features.
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)

# Checking the shape of the scaled data to verify the transformation.
X_scaled.shape

# Splitting the data into training and testing sets (80% train, 20% test).
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# ============================
# Model Selection
# ============================

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (
    RandomForestClassifier,
    GradientBoostingClassifier,
    AdaBoostClassifier,
    BaggingClassifier,
    ExtraTreesClassifier
)
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split the dataset (if not already split)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize models
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "Bagging": BaggingClassifier(),
    "SVM": SVC(probability=True),
    "KNN": KNeighborsClassifier(),
    "MLP": MLPClassifier(max_iter=500),
    "Extra Trees": ExtraTreesClassifier()
}

# Dictionary to store model performance
model_performance = {}

# Train and evaluate each model
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    # Predict on the test set
    y_pred = model.predict(X_test)
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    model_performance[model_name] = accuracy
    print(f"{model_name} Accuracy: {accuracy:.4f}")

# Find the best model
best_model_name = max(model_performance, key=model_performance.get)
best_model_score = model_performance[best_model_name]
best_model = models[best_model_name]

# Print the best model details
print("\n===================")
print(f"Best Model: {best_model_name}")
print(f"Accuracy: {best_model_score:.4f}")
print("===================")

# Example: Printing hyperparameters of the best model
if hasattr(best_model, "get_params"):
    print("\nHyperparameters of the Best Model:")
    for name, value in best_model.get_params().items():
        print(f"{name} -> {value}")
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import RobustScaler

# Initialize variables to store the best fold's metrics
best_accuracy = 0
best_fold_index = None
best_classifier = None
best_predictions = None

# Initialize the Gaussian Naive Bayes classifier
classifier = GaussianNB(priors=None, var_smoothing=1e-09)

# Set the number of folds for cross-validation
n_splits = 11

# Initialize StratifiedKFold for stratified splitting (preserves class distribution)
stratified_kfold = StratifiedKFold(n_splits=n_splits)

# Loop through each fold of cross-validation
for fold_index, (train_index, val_index) in enumerate(stratified_kfold.split(X_scaled, y)):
    # Split the data into training and validation sets for the current fold
    X_train_fold, X_val_fold = X_scaled[train_index], X_scaled[val_index]
    y_train_fold, y_val_fold = y[train_index], y[val_index]

    # Train the model on the current training fold
    classifier.fit(X_train_fold, y_train_fold)

    # Make predictions on the validation set
    y_val_pred = classifier.predict(X_val_fold)

    # Calculate and print the accuracy of the model on the validation set
    accuracy = accuracy_score(y_val_fold, y_val_pred)
    print("Accuracy on fold", fold_index + 1, ":", round(accuracy, 3))

    # Update the best accuracy if the current fold's accuracy is higher
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_fold_index = fold_index
        best_classifier = classifier
        best_predictions = y_val_pred

# Print the best fold index and its corresponding accuracy
print("Best fold index:", best_fold_index + 1)
print("Best accuracy:", round(best_accuracy, 3))

# The best_classifier, best_predictions, and best_accuracy can be used for further analysis or predictions.
```
