```python
import pandas as pd

# Load the Iris dataset from Kaggle
iris_df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')

# Print the first few rows of the dataset
print(iris_df.head())
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot using Matplotlib
plt.scatter(iris_df['sepal_length'], iris_df['sepal_width'])

# Add axis labels and a title
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Dataset - Sepal Length vs Sepal Width')

# Show the plot
plt.show()
```

```python
import seaborn as sns

# Create a scatter plot using Seaborn
sns.scatterplot(x='sepal_length', y='petal_length', data=iris_df)

# Add a title
plt.title('Iris Dataset - Sepal Length vs Petal Length')

# Show the plot
plt.show()
```

```python
import plotly.express as px

# Create a scatter plot using Plotly
fig = px.scatter(iris_df, x='sepal_width', y='petal_length',
                 title='Iris Dataset - Sepal Width vs Petal Length',
                 color="species",
                 size='petal_length',
                 hover_data=['petal_width'])

# Show the plot
fig.show()
```
